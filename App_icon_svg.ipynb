```python
# convert_icon.py
import cairosvg
from PIL import Image
import os

SVG_IN = "app_icon.svg"
PNG_OUT = "app_icon.png"
ICO_OUT = "app_icon.ico"

# 1) Convert SVG -> PNG (large for quality)
cairosvg.svg2png(url=SVG_IN, write_to=PNG_OUT, output_width=1024, output_height=1024)
print("Saved PNG:", PNG_OUT)

# 2) Create .ico with several sizes (for Windows taskbar/start)
img = Image.open(PNG_OUT)
sizes = [(256,256),(128,128),(64,64),(48,48),(32,32),(16,16)]
img.save(ICO_OUT, format='ICO', sizes=sizes)
print("Saved ICO:", ICO_OUT)
```

```python
# proxy_server.py
from flask import Flask, send_from_directory
import threading
import os

from dialogue_app import demo   # Import your Gradio Blocks app

app = Flask(__name__)

@app.route("/")
def home():
    return "✅ Flask WSGI server is running. Visit /app for the Dialogue Creator."

@app.route("/icon")
def icon():
    return send_from_directory(".", "app_icon.svg")

@app.route("/app")
def gradio_link():
    return "Gradio Dialogue Creator running at <a href='http://localhost:7861'>http://localhost:7861</a>"

# Run Gradio in background (on port 7861)
def run_gradio():
    demo.launch(server_name="0.0.0.0", server_port=7861, share=False)

# When imported by Gunicorn, we don’t auto-run Flask
if __name__ == "__main__":
    import threading
    threading.Thread(target=run_gradio, daemon=True).start()
    app.run(host="0.0.0.0", port=7862, debug=False) # Changed port to 7862

# Write the code to a file so gunicorn can find it
with open("proxy_server.py", "w") as f:
    f.write("""
# proxy_server.py
from flask import Flask, send_from_directory
import threading
import os

from dialogue_app import demo   # Import your Gradio Blocks app

app = Flask(__name__)

@app.route("/")
def home():
    return "✅ Flask WSGI server is running. Visit /app for the Dialogue Creator."

@app.route("/icon")
def icon():
    return send_from_directory(".", "app_icon.svg")

@app.route("/app")
def gradio_link():
    return "Gradio Dialogue Creator running at <a href='http://localhost:7861'>http://localhost:7861</a>"

# Run Gradio in background (on port 7861)
def run_gradio():
    demo.launch(server_name="0.0.0.0", server_port=7861, share=False)

# When imported by Gunicorn, we don’t auto-run Flask
if __name__ == "__main__":
    import threading
    threading.Thread(target=run_gradio, daemon=True).start()
    app.run(host="0.0.0.0", port=7862, debug=False) # Changed port to 7862
""")
```
